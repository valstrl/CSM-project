```python
# Classify Images (Output ../result.csv containing all classification results: top 5 guess for each image)
sys.path.append('../')
sys.path.append('./')
from imagenet.classify_image import run_inference_on_image, maybe_download_and_extract, create_graph
import glob

image_list = []
for filename in glob.glob('./images_by_id/*.jpg'):
    image_list.append(filename)
    
f = open('./result.csv', "w")
writer = csv.writer(f)

maybe_download_and_extract()
create_graph()

for idx, image_name in enumerate(image_list):
    image_path = './images_by_id/' + image_name
    human_string, score = run_inference_on_image(image_path)
    print('%d: Image: %s , Class: %s (score = %.5f)' % (idx, image_name,human_string[0], score[0]))
    row = np.concatenate(([image_name],human_string, score))
    writer.writerow(row)

f.close()
print('Finished')
```

```python
## Read groundtruth_hashtags_lang and merge it with images classes and tweet ids

gt_data = pd.read_pickle('./data/groundtruth_hashtags_lang.pkl')

topic_data = pd.read_csv('./topics_for_classifier.csv', header=None, 
                  names = ["id", "tweet_class"])
image_data = pd.read_csv('./imagenet_for_classifier.csv', header=None, 
                  names = ["id", "image_class", "image_score"])

topic_classes = topic_data['tweet_class']
classes = []
for class_ in topic_classes:
    array_class = np.zeros(7)
    array_class[class_] = 1
    classes.append(array_class)

topic_data['t_topic_0'] = np.array(classes)[:,0]
topic_data['t_topic_1'] = np.array(classes)[:,1]
topic_data['t_topic_2'] = np.array(classes)[:,2]
topic_data['t_topic_3'] = np.array(classes)[:,3]
topic_data['t_topic_4'] = np.array(classes)[:,4]
topic_data['t_topic_5'] = np.array(classes)[:,5]
topic_data['t_topic_6'] = np.array(classes)[:,6]
#print(topic_data.head())

image_classes = image_data['image_class']
image_scores = image_data['image_score']

im_classes = []
for idx, class_ in enumerate(image_classes):
    array_class = np.zeros(8)
    array_class[class_] = image_scores[idx]
    im_classes.append(array_class)

image_data['im_class_0'] = np.array(im_classes)[:,0]
image_data['im_class_1'] = np.array(im_classes)[:,1]
image_data['im_class_2'] = np.array(im_classes)[:,2]
image_data['im_class_3'] = np.array(im_classes)[:,3]
image_data['im_class_4'] = np.array(im_classes)[:,4]
image_data['im_class_5'] = np.array(im_classes)[:,5]
image_data['im_class_6'] = np.array(im_classes)[:,6]
image_data['im_class_7'] = np.array(im_classes)[:,7]
#print(image_data.head())

## Merge data 
print(gt_data.shape)
print(topic_data.shape)
print(image_data.shape)
merged_1 = topic_data.merge(image_data, how='outer', left_on='id', right_on='id')
print(merged_1.shape)
merged_2 = gt_data.merge(merged_1, how='inner', left_on='tweet_id', right_on='id')
print(merged_2.shape)
merged_2.to_pickle('data/groundtruth_all.pkl')

'''data_training = merged_2[['Hashtag topic 0', 'Hashtag topic 1', 'Hashtag topic 2', 'Hashtag topic 3', 'Hashtag topic 4',
                             'Hashtag topic 5','Hashtag topic 6','Hashtag topic 7',
                             'retweeted', 
                             'user_lang_fr','user_lang_en','user_lang_de','user_lang_en-gb',
                             'user_lang_ar',
                             'lang_fr','lang_en','lang_und','lang_es','lang_de','other lang','other user lang',
                             't_topic_0','t_topic_1','t_topic_2','t_topic_3','t_topic_4','t_topic_5','t_topic_6',
                             'im_class_0','im_class_1','im_class_2','im_class_3','im_class_4','im_class_5','im_class_6','im_class_7']]'''

#print (data_training.head())
#print (data_label.head())
print ('Done.')
```
